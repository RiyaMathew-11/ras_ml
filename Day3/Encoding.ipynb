```python
import pandas as pd
import sklearn
```

```python
dataset = pd.read_csv('car_models.csv')
dataset
```

### Using pandas to create dummy variables

```python
dummy_val = pd.get_dummies(dataset['Car Model'])
dummy_val
```

```python
merged_dataset = pd.concat([dataset, dummy_val],axis=1)
merged_dataset
```

### Dummy Variable Trap

When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.

<b>NOTE: sklearn library takes care of dummy variable trap hence even if you don't drop one of the state columns it is going to work, however we should make a habit of taking care of dummy variable trap ourselves just in case library that you are using is not handling this for you</b>



```python
final_dataset = merged_dataset.drop(['Car Model', 'Mercedez Benz C class'], axis=1)
```

```python
final_dataset
```

```python
X = final_dataset.drop(['Sell Price($)'], axis=1)
X
```

```python
y = final_dataset['Sell Price($)']
y
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X , y , test_size= 0.1)

from sklearn.linear_model import LinearRegression
my_model = LinearRegression()

my_model.fit(X_train, y_train)

accuracy = my_model.score(X_test,y_test)
accuracy
```

```python
import pickle
with open('model_pickle','wb') as f:
    pickle.dump(my_model, f)
```

### One Hot Encoding

```python
### Use dataset


from sklearn.preprocessing import LabelEncoder
Label_Encode = LabelEncoder()
```

```python
dataset_le = dataset

dataset_le['Car Model'] = Label_Encode.fit_transform(dataset_le['Car Model'])
```

```python
dataset_le
```

```python
y = dataset_le['Sell Price($)']
```

```python
X = dataset_le.drop(['Sell Price($)'], axis=1)
X
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('Car Model', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
```

```python
X
```

```python
X = X[:,1:]
```

```python
X # - scientific notation 
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X , y , test_size= 0.1)

my_model.fit(X_train, y_train)
accuracy = my_model.score(X_test,y_test)
accuracy
```
